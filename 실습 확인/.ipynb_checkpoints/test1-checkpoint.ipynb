```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_exr = pd.read_csv('./data/notExercise.csv')
del df_exr['기간']
del df_exr['대분류']
df_exr
```

```python
df_exr_sex = df_exr[1:3]
df_exr_sex.set_index('분류', inplace = True)
df_exr_sex
```

```python
sexgp = df_exr_sex.plot.pie(y='운동을 할 충분한 시간이 없어서', autopct='%1.1f%%', figsize=(6,6))
sexgp.set_xlabel('성별')
sexgp.set_ylabel('운동을 할 충분한 시간이 없어서')
```
