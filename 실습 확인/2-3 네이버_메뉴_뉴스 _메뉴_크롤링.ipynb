### ÎÑ§Ïù¥Î≤Ñ Îâ¥Ïä§ÏÑπÏÖòÎ©îÎâ¥ÏôÄ ÏÑπÏÖòÎ≥Ñ url Ï∂îÏ∂ú

```python
from urllib.request import urlopen
import bs4
import pandas as pd
```

```python
url = 'https://news.naver.com'

html = urlopen(url)

bs_obj = bs4.BeautifulSoup(html, 'html.parser')
```

```python
# ÎÑ§Ïù¥Î≤Ñ Îâ¥Ïä§ ÏÑπÏÖòÎ©îÎâ¥ ÌÉúÍ∑∏ ÌôïÏù∏(Í∞úÎ∞úÏûêÎèÑÍµ¨)
# body > section > header > div.Nlnb._float_lnb > div > div > div.Nlnb_left._lnb_scroll > div > div > ul
# selectorÍ∞Ä ÎÑàÎ¨¥ Í∏∏Ïñ¥ÏÑú Ïú†Ïö©ÌïòÏßÄ ÏïäÏïÑ Î≥¥ÏûÑ
# ÏßÅÏ†ë ÌôïÏù∏Ìïú ÌÉúÍ∑∏ÏôÄ ÌÅ¥ÎûòÏä§ ÏÜçÏÑ± ÏÇ¨Ïö©
# ul ÌÉúÍ∑∏Ïùò class : Nlnb_menu_list
uls = bs_obj.findAll("ul",{"class":"Nlnb_menu_list"})
len(uls) # ÏõêÏÜåÍ∞Ä 1Í∞úÏù¥ÎØÄÎ°ú
ul = bs_obj.find("ul",{"class":"Nlnb_menu_list"})
ul
```

```python
lis = ul.findAll('li')
len(lis)
```

```python
for li in lis :
    a_tag = li.find('a')
    print(a_tag.text," : ",a_tag['href'])    
```

```python
# liÌÉúÍ∑∏Î•º Ïù¥Ïö©Ìï¥ÏÑú Ï∂îÏ∂ú
lis = bs_obj.findAll("li",{"class" : "Nlist_item"})
len(lis)
```

```python
for li in lis :
    a_tag = li.find('a')
    print(a_tag.text)
    print(a_tag['href'])
```

```python
# ÏàòÏßë Îç∞Ïù¥ÌÑ∞ dfÎ°ú Íµ¨ÏÑ± ÌõÑ Ï†ÄÏû•
section = []
link = []
```

```python
for li in lis :
    a_tag = li.find('a')
    section.append(a_tag.text)
    link.append(a_tag['href'])
```

```python
col_dict = {'section':section, "link":link}
news_section_df = pd.DataFrame(col_dict)
news_section_df
```
