# 변수
 - 자료(data)를 넣을 수 있는 상자를 변수(variable)라 하고 그 상자에 붙인 이름을 변수명(변수 이름)이라고 한다.
 - 실제로 변수는 컴퓨터의 임시 저장 공간(메모리)에 저장된다.
 - 파이썬에서는 등호(=)를 이용해 변수에 자료를 넣는다.(할당)
 - 변수명은 문자, 숫자, 밑줄 기호(_) 등을 이용해 만들고, 대소문자를 구분한다.
 - 변수명에는 공백을 포함할 수 없으며, 밑줄 이외의 기호와 예약어는 변수명으로 이용할 수 없다. 또한 변수명은 숫자로 시작할 수 없다.
 - 변수명을 한글로 지정하는 것 최대한 지양
# 상수
 - 상수는 프로그래밍 언어에서 어떤 숫자를 변수에 할당한 후에 프로그램이 끝날 때 까지 그 변수의 값을 변경하지 않고 사용하는 경우이다.(constant variable)
# 변수 범위
 - 변수의 위치에 따른 차이 : 지역변수, 전역변수

# 문자열
 - 문자열은 문자의 나열을 의미하며, 큰따옴표("), 작은따옴표(')로 둘러싸인 문자의 집합
 - CHAR와 VARCHAR의 차이를 알아두자
 - 컴퓨터는 문자열을 입력할 때 문자열의 끝에 보이지않는 \n을 추가하여 문자열의 끝을 구분한다.
 - 문자열 안에 큰따옴표나 작은따옴표를 포함하고 싶다면 다른 따옴표로 감싸고 포함하고자하는 따옴표 기호를 넣으면 된다.
 - 따옴표 두 종류를 다 포함하고 싶거나 문장을 여러 행을 출력하고 싶다면 문자열 전체를 삼중 따옴표로 감싸면 된다.
 - 문자열에는 더하기 연산자(+)와 곱하기 연산자(*)를 이용할 수 있다.
    - 더하기 연산자는 문자열끼리 연결(concatenation)해 문자열을 하나로 만든다.
    - 곱하기 연산자는 곱한 만큼 문자열을 반복한다.

# 자료형, 집합자료형
 - 리스트(List) - [0, 1, 2, 3, 4, 5]
 - 튜플(tuple) - (0, 1, 2, 3, 4, 5)
 - 세트(set) - {0, 1, 2, 3, 4, 5}
 - 딕셔너리(dictionary) - { key1 : value1, key2 : value2 }
 - 생성(만들기)
 - 데이터 다루기(검색, 추가, 삭제, 변경)
 - 리스트, 튜플 : 순서 O | 세트, 딕셔너리 : 순서 X
 - 리스트, 딕셔너리 : 변경 O | 튜플, 세트 : 변경 X
 - 리스트, 딕셔너리, 세트 : 추가 O | 튜플 : 추가 X

# 리스트
 - 리스트에서 각 항목은, 변수명[i]로 지정할 수 있음, 이를 인덱스(index)라고 한다.
 - 리스트의 인덱스는 0부터 시작한다.
 - 마지막 항목은 변수명[-1]로도 지정이 가능하다.
 - 변수명[i] = new_data 와 같은 형식으로 리스트의 i 항목을 변경할 수 있다.
 - 리스트도 더하기와 곱하기가 가능하다
   - 더하기는 두 리스트를 연결
   - 곱하기는 리스트를 곱한 수 만큼 반복
 - 리스트[start:end:step] 형식으로 슬라이싱 가능 (start,end,step은 전부 생략가능)
 - 항목을 삭제하려면 del 리스트[i]를 이용
 - 항목의 존재 여부를 확인하려면 '항목 in 리스트'를 이용 : 결과값은 True, False인 bool 타입으로 나온다.
 - 리스트 메서드를 활용 : append(), insert(), extend(), remove(), pop(), index(), count(), sort(), reverse()

# 튜플
 - 튜플의 속성은 리스트와 유사하나 한 번 입력(생성)하면 그 이후에는 항목을 수정, 삭제가 불가능하다.
 - 튜플을 생성할 때 항목이 하나라면 항목을 입력한 후 콤마(,)를 입력해야한다.
 - 튜플의 요소를 변경, 삭제는 불가능하나 index(), count() 같이 요소를 변경하지 않는 메서드는 사용 가능하다.

# 세트
 - 세트는 수학의 집합 개념을 파이썬에서 이용할 수 있도록 만든 데이터 타입이다.
 - 데이터의 순서가 없고 데이터를 중복해서 쓸 수 없다.
 - 교집합(intersection), 합집합(union), 차집합(difference)을 구하는 메서드를 사용 가능하다.
 - 각 메서드는 &, |, - 와 같은 연산자로 대신할 수 있다.

## 자료형 간의 타입 변환
 - 여러 데이터를 다루다 보면 연산이나 처리를 할 때 데이터의 타입을 변환해야 할 필요가 있다.
 - 데이터 타입은 list(), tuple(), set()를 이용해서 서로 변환 가능하다.

# 딕셔너리
 - 딕셔너리의 키(key)와 값(value)의 구분은 콜론(:)으로 한다.
 - 지정한 키에 해당하는 값을 얻고 싶다면 dict[key]의 형식으로 찾을 수 있다.
 - 딕셔너리의 키는 숫자와 문자열이 될 수 있고 값은 어떤 데이터 형태든 다 올 수 있다.
 - 생성된 딕셔너리에 새로운 키와 값을 추가하거나 수정하려면 dict[key] = value의 형식으로 입력한다.
 - 특정 키와 값을 삭제하려면 del dict[key] 형식으로 입력한다.
 - 딕셔너리 메서드를 활용 : keys(), values(), items(), update(), clear()
