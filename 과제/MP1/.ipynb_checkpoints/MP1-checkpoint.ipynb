```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='Malgun Gothic')
```

```python
input_data = pd.read_excel("notExercise.xls")
```

```python
input_data
```

```python
data_by_gender = input_data.loc[1:2,"분류":"기타"]
data_by_gender
```

```python
plt.figure(figsize=(8,12))
for i in range(1,len(data_by_gender.columns)):
    plt.subplot(3,2,i)
    plt.pie(data_by_gender.loc[0:2,data_by_gender.columns[i]], labels=data_by_gender.loc[0:2,"분류"], autopct='%.1f%%', explode=(0.03, 0))
    plt.title(data_by_gender.columns[i],fontdict={'fontsize' : 10})
plt.savefig('MP1-1.png', dpi = 200)
```

```python
data_by_age = input_data.iloc[3:9]
data_by_age
```

```python
plt.figure(figsize=(12,8))
for i in range(3,len(data_by_age.columns)):
    plt.subplot(2,3,i-2)
    plt.pie(data_by_age.iloc[:,i], labels=data_by_age.iloc[:,2], autopct='%.1f%%', explode=(0.02, 0.02, 0.02, 0.02, 0.02, 0.02))
    plt.title(data_by_age.columns[i],fontdict={'fontsize' : 10})
plt.savefig('MP1-2.png', dpi = 200)
```

```python
data_by_grade = input_data[input_data["대분류"] == "학력별"]
data_by_grade
```

```python
plt.figure(figsize=(8,20))
for i in range(3,len(data_by_grade.columns)):
    plt.subplot(5,1,i-2)
    plt.pie(data_by_grade[data_by_grade.columns[i]], labels=data_by_grade["분류"], autopct='%.1f%%', explode=(0.02, 0.02, 0.02, 0.02))
    plt.title(data_by_grade.columns[i],fontdict={'fontsize' : 10})
plt.savefig('MP1-3.png', dpi = 200)
```
