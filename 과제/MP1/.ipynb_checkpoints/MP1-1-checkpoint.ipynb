```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
class Draw_Graph():
    
    mpl.rcParams['axes.unicode_minus'] = False
    plt.rc('font', family='Malgun Gothic')
    data = pd.read_excel("notExercise.xls")
    tmp_list = []
    for item in data["대분류"].unique():
        if not item.endswith("구"):
            tmp_list.append(item)
    tmp_list.append("지역소분류")
    category = {i + 1: item for i, item in enumerate(tmp_list)}
    
    def __init__(self):
        pass

    def choose_category(self, cate_num):
        try:
            if self.cate_num == 8:
                self.g_data = data[data["대분류"].str.endswith("구")]
                self.g_data = self.g_data.replace("-",0)
            else:
                self.g_data = data[data["대분류"] == category[cate_num]]
            self.g_data = self.g_data.iloc[:,2:]
            self.draw_pie()
        except:
            print("잘못 누르셨습니다. 올바른 번호를 선택해주세요.")

    def draw_pie(self):
        explode_tuple = (0.015,) * (len(self.g_data.index))
        plt.figure(figsize=(8,12))
        for i in range(1, len(self.g_data.columns)):
            plt.subplot(3,2,i)
            plt.pie(self.g_data.iloc[:,i],labels=self.g_data.loc[:,"분류"],autopct='%.1f%%',explode=explode_tuple)
            plt.title(self.g_data.columns[i])
        plt.suptitle(category[self.cate_num],fontsize=16)
        plt.show()

    def draw_graph(self):
        for key, value in category.items():
            print(f'{key}. {value}')
        self.cate_num = int(input("번호를 선택하세요."))
        self.choose_category(self.cate_num)
```

```python
dg = Draw_Graph()
dg.draw_graph()
```
